## Grafico de dispersion (posicion de salida vs posicion final)

```python
plt.figure(figsize=(10,6))
plt.scatter(df_results['grid'], df_results['positionOrder'], alpha=0.5)
plt.title('Posición de salida vs posición de llegada')
plt.xlabel('Posición de salida (grid)')
plt.ylabel('Posición de llegada (positionOrder)')
plt.show()
```

```python
plt.figure(figsize=(30,20))
df_results.boxplot(column='positionOrder', by='constructorId', grid=False)
plt.title('Boxplot de posición de llegada por equipo (constructorId)')
plt.suptitle('')
plt.xlabel('ID de equipo (constructorId)')
plt.ylabel('Posición de llegada')
plt.show()
```

```python
correlation_matrix = df_results.corr(numeric_only=True)
correlation_matrix
```

```python
plt.figure(figsize=(6,4))  # tamaño del gráfico
sns.heatmap(correlation_matrix, 
            annot=True,        # mostrar valores
            cmap="coolwarm",   # esquema de colores
            vmin=-1, vmax=1)   # rango de correlación
plt.title("Matriz de Correlación")
plt.show()
```
